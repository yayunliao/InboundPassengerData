```python
import pandas as pd
import numpy as np
import datetime
import os
import re
```

```python
# 來臺旅客按年齡及來臺目的交叉分析
df = pd.read_csv('新/交叉/來臺旅客按年齡及來臺目的交叉分析.csv', encoding="utf-8")
df
```

```python
# column name
new_columns_name = df.iloc[0]
df.columns = new_columns_name

# 刪除第一row、最後一row
df = df.iloc[1:-1]

# 改資料型態
df['Years'] = df['Years'].astype(int)
df['Years'] = df['Years'] + 1911

# Dates
df['Dates'] = pd.to_datetime(df['Years'].astype(str) + '-' + df['Months'].astype(str), format=('%Y-%m'))
df = df.drop(['Years', 'Months'], axis=1)

# 將'Dates'欄位移到最前面
df = df[['Dates'] + df.columns[:-1].tolist()]

df['Purpose'] = df['Purpose'].str.extract(r'([a-zA-Z\sa-zA-Z]+)')

df.to_csv('新整理/交叉/來臺旅客按年齡及來臺目的交叉分析.csv', encoding='utf-8', index=False)
df
```

```python
# 來臺旅客按來臺目的及性別及年齡交叉分析
df_2 = pd.read_csv('新/交叉/來臺旅客按來臺目的及性別及年齡交叉分析.csv', encoding='utf-8')
df_2
```

```python
# 第一列中的缺失值使用前一個非缺失值來進行填充
df_2.iloc[0].fillna(method='ffill', inplace=True)

# columns name
df_2.iloc[2,3:] = df_2.iloc[0,3:] + '_' + df_2.iloc[2,3:]
new_columns_name = df_2.iloc[2]
df_2.columns = new_columns_name

# 刪除前 3 row、最後一 row
df_2 = df_2.iloc[3:-1]

# 改資料型態
df_2['Years'] = df_2['Years'].astype(int)
df_2['Years'] = df_2['Years'] + 1911

# Dates
df_2['Dates'] = pd.to_datetime(df_2['Years'].astype(str) + '-' + df_2['Months'].astype(str), format=('%Y-%m'))
df_2 = df_2.drop(['Years', 'Months'], axis=1)

# 將'Dates'欄位移到最前面
df_2 = df_2[['Dates'] + df_2.columns[:-1].tolist()]

# Purpose 去除中文
pattern = r'\((.*?)\)'  # 匹配最後一對小括號內的文字

def extract_text(row):
    matches = re.findall(pattern, row)
    if matches:
        return matches[-1]
    else:
        return ""

df_2['Purpose'] = df_2['Purpose'].str.replace('\n', '')  # 移除換行符號
df_2['Purpose'] = df_2['Purpose'].apply(extract_text)

df_2.to_csv('新整理/交叉/來臺旅客按來臺目的及性別及年齡交叉分析.csv', encoding='utf-8', index=False)

df_2
```

```python

data_2 = pd.read_csv('新整理/交叉/來臺旅客按來臺目的及性別及年齡交叉分析.csv')

# 初始化新的資料框 new_df，該資料框將用於儲存重組後的資料
new_df = pd.DataFrame(columns=['Dates', 'Purpose', 'Age', 'Male', 'Female'])

# Iterate over the rows in the original DataFrame
for _, row in data_2.iterrows():
    # Iterate over age groups and genders
    for age_group, gender in [('1-12Years', 'Male'), ('1-12Years', 'Female'),
                              ('13-19Years', 'Male'), ('13-19Years', 'Female'),
                              ('20-29Years', 'Male'), ('20-29Years', 'Female'),
                              ('30-39Years', 'Male'), ('30-39Years', 'Female'),
                              ('40-49Years', 'Male'), ('40-49Years', 'Female'),
                              ('50-59Years', 'Male'), ('50-59Years', 'Female'),
                              ('60Years and Over', 'Male'), ('60Years and Over', 'Female')]:
        # Extract the count value from the corresponding column
        count = row[f'{age_group}_{gender}']
        
        # Append a new row to the new DataFrame
        new_row = {
            'Dates': row['Dates'],
            'Purpose': row['Purpose'],
            'Age': age_group,
            'Male': count if gender == 'Male' else None,
            'Female': count if gender == 'Female' else None
        }
        new_df = new_df.append(new_row, ignore_index=True)

# Combine rows with the same 'Dates', 'Purpose', and 'Age' values
new_df = new_df.groupby(['Dates', 'Purpose', 'Age'], as_index=False).sum()

new_df.to_csv('新整理/交叉/new來臺旅客按來臺目的及性別及年齡交叉分析.csv', index=False, encoding="UTF-8")
```

```python
# 來臺旅客按居住地及來臺目的及性別交叉分析
df_3 = pd.read_csv('新/交叉/來臺旅客按居住地及來臺目的及性別交叉分析.csv', encoding='utf-8')
df_3
```

```python
# 第一列中的缺失值使用前一個非缺失值來進行填充
df_3.iloc[0].fillna(method='ffill', inplace=True)
# columns name
df_3.iloc[2,3:] = df_3.iloc[0,3:] + '_' + df_3.iloc[2,3:]
new_columns_name = df_3.iloc[2]
df_3.columns = new_columns_name

# 刪除前 3 row
df_3 = df_3.iloc[3:]
# 刪除最後一 row
df_3 = df_3.iloc[:-1]

# 改資料型態
df_3['Months'] = df_3['Months'].astype(float).astype(int)
df_3['Years'] = df_3['Years'].astype(int)
df_3['Years'] = df_3['Years'] + 1911

# Dates
df_3['Dates'] = pd.to_datetime(df_3['Years'].astype(str) + '-' + df_3['Months'].astype(str), format=('%Y-%m'))
df_3 = df_3.drop(['Years', 'Months'], axis=1)

# 將'Dates'欄位移到最前面
df_3 = df_3[['Dates'] + df_3.columns[:-1].tolist()]

pattern = r'\((.*?)\)'  # 匹配最後一對小括號內的文字

def extract_text(row):
    matches = re.findall(pattern, row)
    if matches:
        return matches[-1]
    else:
        return ""

df_3['Residence'] = df_3['Residence'].str.replace('\n', '')  # 移除換行符號
df_3['Residence'] = df_3['Residence'].apply(extract_text)

# 刪除Asia、Americas、Oceania、Europe、Africa
df_3 = df_3[df_3['Residence'] != 'Asia']
df_3 = df_3[df_3['Residence'] != 'Americas']
df_3 = df_3[df_3['Residence'] != 'Oceania']
df_3 = df_3[df_3['Residence'] != 'Europe']
df_3 = df_3[df_3['Residence'] != 'Africa']

df_3.to_csv('新整理/交叉/來臺旅客按居住地及來臺目的及性別交叉分析.csv', encoding='utf-8', index=False)

df_3
```

```python
# 重新組合
data_3 = pd.read_csv('新整理/交叉/來臺旅客按居住地及來臺目的及性別交叉分析.csv')
# 初始化新的資料框 new_df，該資料框將用於儲存重組後的資料
new_df = pd.DataFrame(columns=['Dates', 'Residence', 'Purpose', 'Male', 'Female'])

# 逐行迭代原始資料
for _, row in data_3.iterrows():
    # 逐個迭代不同的來臺目的和性別組合
    for purpose_group in ['Business', 'Leisure', 'Visit Relatives', 'Conference', 'Study', 'Exhibition', 'Medical Treatment', 'Others']:
        # 從原始資料框 data_3 中根據來臺目的和性別組合提取對應的計數值
        male_count = row[f'{purpose_group}_Male']
        female_count = row[f'{purpose_group}_Female']
        
        # 創建新的一行數據 new_row，將計數值和相應的日期、居住地和來臺目的填入
        new_row = {
            'Dates': row['Dates'],
            'Residence': row['Residence'],
            'Purpose': purpose_group,
            'Male': male_count,
            'Female': female_count
        }
        # 將新的一行數據加入到新的資料框 new_df 中
        new_df = new_df.append(new_row, ignore_index=True)

new_df.to_csv('新整理/交叉/new來臺旅客按居住地及來臺目的及性別交叉分析.csv', index=False, encoding='utf-8')
new_df
```
