```python
import pandas as pd
import numpy as np
from datetime import datetime
import os
import re
```

```python
# 輸出資料的資料夾
out_folder_name = '新整理_民宿住客類別及國籍人數統計表'
out_folder_path = os.path.join('final_report', out_folder_name)
os.makedirs(out_folder_path, exist_ok=True)
```

```python
# 輸入資料的資料夾
fold_path = '民宿住客類別及國籍人數統計表_csv'
# 取得資料夾內所有檔案名稱
files_name = os.listdir(fold_path)

for file_name in files_name:
    if file_name.endswith('.csv'):
        # 組成檔案路徑
        file_path = os.path.join(fold_path, file_name)
        
        # 讀取csv
        df = pd.read_csv(file_path)
        
        # 填滿 row1
        df.iloc[1].fillna(method='ffill', inplace=True)
        df.iloc[1,1:3] = df.iloc[1,1:3] + '_' + df.iloc[2,1:3]
        df.iloc[1,4:17] = df.iloc[1,4:17] + '_' + df.iloc[2,4:17]
        df.iloc[1,18:] = df.iloc[1,18:] + '_' + df.iloc[2,18:]
        
        # new column name
        new_column_name = df.iloc[1]
        df.columns = new_column_name
        
        # 新增column 日期
        regex01 = r'\d+年\d+月'
        string01 = df.iloc[0, 0]
        match = re.search(regex01, string01)
        if match:
                date_value = match.group()
                df['日期'] = date_value
                
        # 移除不要的row
        df = df.drop(df.index[[0,1,2,-1]])
        
        # 改日期格式
        date_str = str(df.iloc[0,-1])
        date_obj = datetime.strptime(date_str, "%Y年%m月")
        formatted_date = date_obj.strftime("%Y/%m/%d")
        df['日期'] = formatted_date
        
        # 日期column 移到最前面
        date_column = df['日期']
        df = df.drop('日期', axis=1)
        df.insert(0, '日期', date_column)
        
        # 改column名稱
        df.rename(columns={'縣市/項目': '縣市'}, inplace=True)
        df.rename(columns={'住宿人次合計': '住宿人數合計'}, inplace=True)

        # 需要删除的column
        columns_to_delete = ['住宿人數合計','接待外籍旅客家數(本國以外)', '員工人數_客房部', '員工人數_餐飲部', '員工人數_管理部', '員工人數_其他部門', '員工人數_合計']
        # 删除指定的column
        df = df.drop(columns=columns_to_delete)
        
        # 改column名稱
        df.rename(columns={'住宿國籍_大陸': '住宿國籍_中國大陸'}, inplace=True)

        # 儲存到 新整理hotel
        out_file_name = file_name.replace('.csv', '_processed.csv')
        out_file_path = os.path.join(out_folder_path, out_file_name)
        df.to_csv(out_file_path, index=False)
```

```python
# 輸入資料的資料夾
fold_path = '民宿住客類別及國籍人數統計表_csv/2021年以後'
# 取得資料夾內所有檔案名稱
files_name = os.listdir(fold_path)

for file_name in files_name:
    if file_name.endswith('.csv'):
        # 組成檔案路徑
        file_path = os.path.join(fold_path, file_name)
        # 讀取csv
        df = pd.read_csv(file_path)
        # 填滿 row1
        df.iloc[2].fillna(method='ffill', inplace=True)
        df.iloc[2,1:3] = df.iloc[2,1:3] + '_' + df.iloc[3,1:3]
        df.iloc[2,4:30] = df.iloc[2,4:30] + '_' + df.iloc[3,4:30]
        df.iloc[2,31:-1] = df.iloc[2,31:-1] + '_' + df.iloc[3,31:-1]
        
        # new column name
        new_column_name = df.iloc[2]
        df.columns = new_column_name
        
        # 新增column 日期
        regex01 = r'\d+年\d+月'
        string01 = df.iloc[1, 0]
        match = re.search(regex01, string01)
        if match:
                date_value = match.group()
                df['日期'] = date_value
                
        # 移除不要的row
        df = df.drop(df.index[[0,1,2,3,-1]])
        
        # 改日期格式
        date_str = str(df.iloc[0,-1])
        date_obj = datetime.strptime(date_str, "%Y年%m月")
        formatted_date = date_obj.strftime("%Y/%m/%d")
        df['日期'] = formatted_date
        
        # 日期column 移到最前面
        date_column = df['日期']
        df = df.drop('日期', axis=1)
        df.insert(0, '日期', date_column)
        
        # 改column名稱
        df.rename(columns={'縣市/項目': '縣市'}, inplace=True)
        
        # 新增'住宿國籍_其他亞洲地區'、'住宿國籍_歐洲'、'住宿國籍_北美'
        df['住宿國籍_其他亞洲地區'] = df['住宿國籍_泰國'].astype(float) + df['住宿國籍_印尼'].astype(float) + df['住宿國籍_越南'].astype(float) + df['住宿國籍_菲律賓'].astype(float) + df['住宿國籍_汶萊'].astype(float) + df['住宿國籍_緬甸'].astype(float) + df['住宿國籍_寮國'].astype(float) + df['住宿國籍_柬埔寨'].astype(float)+ df['住宿國籍_印度'].astype(float)
        df['住宿國籍_歐洲'] = df['住宿國籍_英國'].astype(float) + df['住宿國籍_歐洲其他地區'].astype(float)
        df['住宿國籍_北美'] = df['住宿國籍_美國'].astype(float) + df['住宿國籍_加拿大'].astype(float)
        
        # 需要删除的column
        columns_to_delete = ['住宿人次合計', '住宿國籍_英國', '住宿國籍_歐洲其他地區', '接待外籍旅客家數(本國以外)', '登記經營者_男性', '登記經營者_女性', '登記經營者_小計', '員工人數_男性', '員工人數_女性', '員工人數_小計', '經營者及員工人數總計', '住宿國籍_英國', '住宿國籍_歐洲其他地區','住宿國籍_美國','住宿國籍_加拿大','住宿國籍_泰國','住宿國籍_印尼','住宿國籍_越南','住宿國籍_菲律賓','住宿國籍_汶萊','住宿國籍_緬甸','住宿國籍_寮國','住宿國籍_柬埔寨','住宿國籍_印度']

        # 删除指定的column
        df = df.drop(columns=columns_to_delete)
        
        # 改column名稱
        df.rename(columns={'住宿國籍_南韓': '住宿國籍_韓國'}, inplace=True)

        # 儲存到 新整理hotel
        out_file_name = file_name.replace('.csv', '_processed.csv')
        out_file_path = os.path.join(out_folder_path, out_file_name)
        df.to_csv(out_file_path, index=False)
```
