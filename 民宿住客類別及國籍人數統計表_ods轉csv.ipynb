```python
import pandas as pd
import os
from pandas_ods_reader import read_ods
```

```python
folder_name = '民宿住客類別及國籍人數統計表_csv'
os.makedirs(folder_name, exist_ok=True)
```

```python
# 105年
# 讀.ods檔
data = pd.read_excel('民宿住客類別及國籍人數統計表/sheet/105年4-7月民宿住客類別及國籍人數統計表.ods', engine='odf', sheet_name=None)

# 每個sheet儲存為csv
for sheet_name, sheet_data in data.items():
    # csv路徑
    csv_file_path = os.path.join(folder_name, f'{sheet_name}.csv')
    # 儲存為csv
    sheet_data.to_csv(csv_file_path, index=False)
    print(f'Saved {csv_file_path} successfully.')
```

```python
# 2020年前資料

# 資料夾路徑
folder_path = '/Users/liaoyayun/Documents/final_report/民宿住客類別及國籍人數統計表'
# 輸出資料的資料夾路徑
out_folder_path = '/Users/liaoyayun/Documents/final_report/民宿住客類別及國籍人數統計表_csv'
# 讀取資料夾中的檔案列表
file_names = os.listdir(folder_path)

for file_name in file_names:
    if file_name.endswith('.ods'):
        # 組成檔案路徑
        file_path = os.path.join(folder_path, file_name)
        # 讀.ods檔
        data = pd.read_excel(file_path, engine='odf', sheet_name=None)
        
        for sheet_name, sheet_data in data.items():
            csv_file_name = f'{os.path.splitext(file_name)[0]}_{sheet_name}.csv'
            csv_file_path = os.path.join(out_folder_path, csv_file_name)
            
            # 儲存為csv
            sheet_data.to_csv(csv_file_path, index=False)
            print(f'Saved {csv_file_name} successfully.')
```
